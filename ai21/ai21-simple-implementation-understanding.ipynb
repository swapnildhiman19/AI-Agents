```python
import os
from dotenv import load_dotenv

load_dotenv()
AI21_API_KEY = os.getenv("AI21_API_KEY")
```

```python
from ai21 import AI21Client
from ai21.models.chat import ChatMessage

client = AI21Client(
    # defaults to os.environ.get('AI21_API_KEY')
    api_key= AI21_API_KEY,
)

system = "You're one of the most powerful AI in the world, who has all the information about what's happening in the world and how will AI impact the world, how it will eat existing jobs in Software Engineering"
messages = [
    ChatMessage(content=system, role="system"),
    ChatMessage(content="I recently heard that OpenAI is going to make a new OS like how we have Windows and iOS, which will be totally AI powered. Now if this happened than all the jobs like mine (iOS Swift Developer at Walmart), will be finished, because no one would be requiring the Apps", role="user"),
]

chat_completions = client.chat.completions.create(
    messages=messages,
    model="jamba-mini",
)
```

```python
print(chat_completions.choices[0].message.content)
```
